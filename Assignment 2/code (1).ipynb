# Task 1: Input & Data Representation

## To Input:
* Number of processes
* Number of resources
* Allocation matrix
* Maximum matrix
* Available vector

#### 1.1 Basic Input

````python
n = int(input("Enter number of processes: "))
m = int(input("Enter number of resources: "))
````

#### 1.2 Input Allocation Matrix

````python
allocation = []
print("\nEnter Allocation Matrix:")

for i in range(n):
    row = list(map(int, input(f"P{i}: ").split()))
    allocation.append(row)


print("\nAllocation Matrix:")
print("-" * 30)
for i in range(n):
    print(f"P{i}: {allocation[i]}")
````

#### 1.3 Input Maximum Matrix

````python
maximum = []
print("\nEnter Maximum Matrix:")

for i in range(n):
    row = list(map(int, input(f"P{i}: ").split()))
    maximum.append(row)


print("\nMaximum Matrix:")
print("-" * 30)
for i in range(n):
    print(f"P{i}: {maximum[i]}")
````

#### 1.4 Input Available Resources

````python
available = list(map(int, input("\nEnter Available Resources: ").split()))
````

````python
print("\nAvailable Resources:")
print("-" * 25)
for i, val in enumerate(available):
    print(f"R{i}: {val}")
````

# Task 2: Matrix Calculation

### Formula:

* Need = Maximum - Allocation

#### 2.1 Implementation

````python
need = []

for i in range(n):
    row = []
    for j in range(m):
        value = maximum[i][j] - allocation[i][j]
        row.append(value)
    need.append(row)
````

#### 2.2 Display Need Matrix

````python
print("\nNeed Matrix:")
print("-" * 30)

for i in range(n):
    print(f"P{i}: {need[i]}")
````

# Task 3: Banker’s Safety Algorithm

### Logic:

* Work = Available
* Finish = False for all
* Find process where:
```Need <= Work```
* Execute it → update Work
* Repeat

#### 3.1 Implementation

````python
finish = [False] * n 
safe_sequence = []           
work = available.copy()

while len(safe_sequence) < n:
    found = False

    for i in range(n):
        if not finish[i]:
            
            if all(need[i][j] <= work[j] for j in range(m)):

                print(f"\nProcess P{i} is executing...")

                for j in range(m):
                    work[j] += allocation[i][j]

                safe_sequence.append(i)
                finish[i] = True
                found = True

                print(f"Updated Available (Work): {work}")

    if not found:
        break
````

# Task 4: Safe Sequence

#### 4.1 Check System State

````python
if len(safe_sequence) == n:
    print("\nSystem is in SAFE state")
else:
    print("\nSystem is in UNSAFE state (Deadlock possible)")
````

#### 4.2 Safe Sequence Determination

````python
if len(safe_sequence) == n:
    print("\nSafe Sequence:")
    print("-" * 30)
    print(" -> ".join([f"P{i}" for i in safe_sequence]))
````

# Task 5: Result Analysis

### If SAFE:

* All processes executed
* No deadlock
* Sequence exists
    

### If UNSAFE:

* Some processes cannot proceed
* Deadlock possible

#### 5.1 Check

````python
if len(safe_sequence) == n:
    print("\nSystem Status:")
    print("-" * 30)
    print("System is in SAFE state (No Deadlock)")
else:
    print("\nSystem Status:")
    print("-" * 30)
    print("System is in UNSAFE state (Deadlock may occur)")
````

#### 5.2 Detailed Analysis

````python
if len(safe_sequence) == n:
    print("\nAnalysis:")
    print("-" * 30)
    print("✔ All processes were able to execute successfully.")
    print("✔ Resources were allocated and released safely.")
    print("✔ No deadlock situation occurred.")
    print("✔ A safe sequence exists, ensuring system stability.")
else:
    print("\nAnalysis:")
    print("-" * 30)
    print("✘ Not all processes could be executed.")
    print("✘ Some processes are waiting indefinitely.")
    print("✘ Deadlock condition may occur.")
    print("✘ No safe sequence exists.")
````
